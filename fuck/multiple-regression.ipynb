# 다중 회귀분석

```python
import pandas as pd
import numpy as np

raw_data = {
    "EMP_NO": [1, 2, 3, 4, 5, 6],
    "NAME": ["Aurora", "Bill", "Charlie", "Dragon", "Entropy", "False"],
    "AGE": [20, 24, 22, 30, 52, 14],
    "SALARY": [1000, 3000, 2000, 5000, 10000, 100],
    "DEPARTMENT": ["IT", "IT", "HR", "IT", "CEO", "HR"]
}

df = pd.DataFrame(raw_data)
df = df.drop(columns=["NAME", "DEPARTMENT"])
df.head(5)
```

```python
from patsy import dmatrices
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

y, x = dmatrices("SALARY ~ AGE + EMP_NO", data=df, return_type="dataframe")

df_vif = pd.DataFrame()
df_vif["colname"] = x.columns
df_vif["VIF"] = [vif(x.values, i) for i in range(x.shape[1])]
df_vif
```
